```python
from selenium import webdriver

import re
import time

driver = webdriver.Chrome('./chromedriver')
driver.implicitly_wait(3)

driver.get('https://kin.naver.com/people/expert/index.nhn')
```

```python
def get_page_list(class_name, n):
    all_links = driver.find_elements_by_tag_name("a")

    for links in all_links:
        if "page" in links.get_attribute("href"):
            link = links.get_attribute("href")

    syntax = link[:link.rfind("=")+1]

    summary = driver.find_element_by_class_name(class_name).text
    all_exp = int(re.findall(r'\d+', summary)[0])

    count = all_exp // n
    
    if all_exp % n != 0:
        count += 1

    page_list = []

    for i in range(1, count + 1):
        page_list.append(syntax + "{0}".format(i))

    return page_list[:3]
```

```python
exp_tab = driver.find_element_by_class_name("exp_tab")

tab_links = []
for col in exp_tab.find_elements_by_tag_name("a"):
    tab_links.append(col.get_attribute("href"))
    
for col in tab_links:
    driver.get(col)
    print("Tab", col)
    
    exp_page_list = get_page_list("list_summary", 10)
    
    for exp_page in exp_page_list:
        driver.get(exp_page)
        print("exp:", exp_page)
    
        exp_list = driver.find_element_by_class_name("pro_list")

        exp_links = []
        for exp in exp_list.find_elements_by_tag_name("li"):
            for link_list in exp.find_elements_by_tag_name("a"):
                tmp = link_list.get_attribute("href")

                if "answer" in tmp:
                    exp_links.append(tmp)

        for exp in exp_links:
            driver.get(exp)
            
            con_page_list = get_page_list("mykin_num", 20)
            
            for con_page in con_page_list:
                driver.get(con_page)
                print("con:", con_page)

                q_list = driver.find_element_by_id("au_board_list")
                content_list = q_list.find_elements_by_class_name("title")[:3]

                con_links = []
                for content in content_list:
                    con_links.append(content.find_element_by_tag_name("a").get_attribute("href"))

                for con_link in con_links:

                    try:  # 비공개 게시물이 에러를 야기
                        driver.get(con_link)

                        q_area = driver.find_element_by_class_name("question-content")
                        q_title = q_area.find_element_by_class_name("title").text
                        print(q_title)

                        for i in q_area.find_elements_by_class_name("c-heading__content"):
                            print(i.text)

                        ans_list = driver.find_element_by_class_name("se-main-container")

                        for ans in ans_list.find_elements_by_tag_name("span"):
                            print(ans.text)

                        driver.back()

                    except:
                        driver.back()
                
            time.sleep(10)
```
